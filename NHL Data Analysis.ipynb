```python
# Combine the data 

df = pd.concat([year0,year1,year2,year3,year4,year5,year6,year7,year8,year9,year10,year11,year12,year13])
df
```

```python
#The data has situations: all, 5v5, 5v4, 4v5, 4v4, other
#Filter out any data not with the situation being all

df = df[df.situation == 'all']

df
```

```python
#Only select the columns needed for analysis

df = df.filter(['team','season','goalsFor','goalsAgainst','reboundsFor','reboundsAgainst',
'shotsOnGoalFor','shotsOnGoalAgainst','savedShotsOnGoalFor','savedShotsOnGoalAgainst',
'penalityMinutesFor','penalityMinutesAgainst','faceOffsWonFor','faceOffsWonAgainst',
'hitsFor','hitsAgainst','giveawaysFor','giveawaysAgainst'], axis=1)
df
```

```python
#Check for missing data (null)

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} : {}%'.format(col, round(pct_missing*100)))
    
#No data is missing (nothing is empty, could still not have data)
```

```python
#Check for duplicate values

duplicate_values = df.duplicated()
for i in duplicate_values:
    if i == True:
        print(i) 
        
#There are no dupicate values
```

```python
#Combine the data sets on Team to get cumulative stats
#Doing so notice that some teams do not have compleete information given their abnormally low stats

aggregation_functions = {'team': 'first', 'goalsFor': 'sum','goalsAgainst': 'sum', 'reboundsFor': 'sum','reboundsAgainst': 'sum',
                         'shotsOnGoalFor': 'sum','shotsOnGoalAgainst': 'sum','savedShotsOnGoalFor': 'sum','savedShotsOnGoalAgainst': 'sum',
                         'penalityMinutesFor': 'sum','penalityMinutesAgainst': 'sum','faceOffsWonFor': 'sum','faceOffsWonAgainst': 'sum',
                        'hitsFor': 'sum','hitsAgainst': 'sum','giveawaysFor': 'sum','giveawaysAgainst': 'sum'}
df_team = df.groupby(df['team']).aggregate(aggregation_functions)
```

```python
df_team
```

```python
#Sort the values to see who scored the most

df_team = df_team.sort_values(by=['goalsFor'],inplace=False, ascending=False)
df_team
```

```python
#Visualize with a bar chart the top scoring teams

df_team.plot.bar(x='team',y='goalsFor')
```

```python
#Let's compare the Ottawa Senators and Toronto Maple Leafs 

df_ottawa = df.loc[df['team'] == 'OTT']
df_toronto = df.loc[df['team'] == 'TOR']
```

```python
df_ottawa
```

```python
df_toronto
```

```python
#Add a prefix to column names to tell them appart

for col in df_ottawa.columns:
    if col != 'season':
        df_ottawa = df_ottawa.rename(columns={col : 'ottawa_'+col})
for col in df_toronto.columns:
    if col != 'season':
        df_toronto = df_toronto.rename(columns={col : 'toronto_'+col})
```

```python
#Combine the two dataframes

df_ottaronto = pd.concat([df_ottawa,df_toronto])

df_ottaronto
```

```python
#Now we can plot Ottawa and Toronto against eachother over the seasons

df_1 = df_ottaronto.filter(['season','ottawa_goalsFor','ottawa_goalsAgainst','toronto_goalsFor','toronto_goalsAgainst'], axis=1)

df_1.plot.line(x='season',)
```
